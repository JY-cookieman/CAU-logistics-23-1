# Data Exploration

Import packages

```python
# jupyter-notebook-specific command:
# shows the plots without `plt.show()` in Jupyter notebook
%matplotlib inline

# import packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# increase default figure and font sizes for easier viewing
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 14
```

```python
data = pd.read_csv('titanic.csv')
data.head()
```

Rename the class `Survived` with `Class`

```python
data.columns = ['Passenger Class', 'Name', 'Sex', 'Age', 'No of Siblings or Spouses on Board', 'No of Parents or Children on Board', 'Ticket Number', 'Passenger Fare', 'Cabin', 'Port of Embarkation', 'Life Boat', 'Class']
data.head()
```

# 1. Summary Statistics
**1.** For each quantitative attribute, calculate its average, standard deviation, minimum, and maximum values.

```python
for col in ['Age', 'No of Siblings or Spouses on Board', 'No of Parents or Children on Board', 'Passenger Fare']:
  print('%s:' % (col))
  print('\t Mean = %.2f' % data[col].mean())
  print('\t Standard deviation = %.2f' % data[col].std())
  print('\t Minimum = %.2f' % data[col].min())
  print('\t Maximum = %.2f' % data[col].max())
```

As such, you can prepare a list with elements that are quantitative features, including 'Age', 'No of Siblings or Spouses on Board', 'No of Parents or Children on Board', 'Passenger Fare'.

**2.** For the qualitative attributes (classes), count the frequency for each of its distict values.

**<font color="red">Code:</font>**

```python
for col in ['Passenger Class', 'Sex', 'Cabin', 'Port of Embarkation', 'Class']:
  print('%s:' % (col))
  print(data[col].value_counts())
  print()
```

**3.** It is also possible to display the summary for all the attributes simultaneously in a table using the describe() function. If an attribute is quantitative, it will display its mean, standard deviation and various quantiles (including minimum, median, and maximum) values. If an attribute is qualitative, it will display its number of unique values and the top (most frequent) values.

**<font color="red">Code:</font>**

```python
data.describe(include='all')
```

Note that count refers to the number of non-missing values for each attribute.

**4.** For multivariate statistics, you can compute the covariance and correlation between pairs of attributes.

**<font color="red">Code:</font>**

```python
print('Correlation:')
data.corr()
```

Or create subset of original dataframe with only numerical attributes. 
**<font color="red">Code:</font>**

```python
nu_data = data[['Age', 'No of Siblings or Spouses on Board', 'No of Parents or Children on Board', 'Passenger Fare']]
print('Correlation:')
nu_data.corr()
```

## 2. Data Visualization

Data visualization is the display of information in a graphic or tabular format. Successful visualization requires that the data (information) be converted into a visual format so that the characteristics of the data and the relationships among data items or attributes can be analyzed or reported.

In this tutorial, you will learn how to display the Iris data. To execute the sample program shown here, make sure you have installed the `matplotlib` library package.

Befor visualization, you need to import the `matplotlib` package.

**1.** First, we will display the histogram for the `Age` attribute by discretizing it into 8 separate bins and counting the frequency for each bin.

**<font color="red">Code:</font>**

```python
plt.hist(data['Age'], bins=8)
# data['sepal width'].hist(bins=8) # do the same thing as above
plt.xlabel('Age')
plt.ylabel('count')
plt.title('Histogram of Age')
plt.show()
```

**2.** A boxplot can also be used to show the distribution of values for each attribute.

**<font color="red">Code:</font>**

```python
plt.boxplot(nu_data)
plt.xlabel('column')
plt.ylabel('value')
plt.title('Box plot of Titanic')
plt.show()
```

There are missing values in the `Age` and `Passenger Fare` columns of the Titanic data set. If missing value(s) is included, as we can see boxplots will not be created (for the first (`Age`) and fourth (`Passenger Fare`) attributes).

To resolve this problem, either remove those rows, or fill the values with a default before creasing a boxplot. In the code below, fill `Nan` with `0.0` using the corresponding code `.fillna(0.0)`

**2.1.** Below, a new dataframe `nu_data_fill` is created and corresponding boxflots are developed.

```python
nu_data_fill = nu_data.fillna(0.0)
plt.boxplot(nu_data_fill)
plt.xlabel('column')
plt.ylabel('value')
plt.title('Box plot of Titanic')
plt.show()
```

**2.2.** Or drop the rows where at least one element is missing, using `.dropna()`
The code below, create a new dataframe `nu_data_wona` (without Nan).

```python
nu_data_wona = nu_data.dropna()
plt.boxplot(nu_data_wona)
plt.xlabel('column')
plt.ylabel('value')
plt.title('Box plot of Titanic')
plt.show()
```

The results are different.

**3.** For each pair of attributes, we can use a scatter plot to visualize their joint distribution.

```python
x = 'Age'
y = 'Passenger Fare'

plt.scatter(nu_data[x], nu_data[y])

plt.xlabel(x)
plt.ylabel(y)
plt.title('Scatter plot')
plt.show()
```

```python
# Scatter matrix
fig, axes = plt.subplots(4, 4, figsize=(12, 12)) # 4 x 4 plots in a figure

for i in range(4):
    col_i = nu_data.columns[i]
    for j in range(4):
        col_j = nu_data.columns[j]
        if j == i:
            axes[i][j].hist(nu_data[col_i], bins=10)
            axes[i][j].set_xlabel(col_i)
            axes[i][j].set_ylabel('count')
        else:
            axes[i][j].scatter(nu_data[col_i], data[col_j])
            axes[i][j].set_xlabel(col_i)
            axes[i][j].set_ylabel(col_j)
plt.tight_layout()
```

You can obtain the same result with only one line below:
